```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Lista do przechowywania wszystkich danych
all_data = []

# Pętla iterująca przez 4 strony
for page in range(1, 5):
    url = f"https://www.viamichelin.pl/web/Stacje-benzynowe/Stacje-benzynowe-Malopolskie-Polska?page={page}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")

    # Znajdowanie elementów stacji paliw na stronie
    stations = soup.find_all("li", class_="poi-item poi-item-gasStation")

    for station in stations:
        name = station.find("div", class_="poi-item-name truncate").text.strip()

        # Pobieranie adresu
        address_div = station.find("div", class_="poi-item-details-address truncate")
        address = address_div.text.strip() if address_div else "Brak adresu"

        # Wyodrębnienie kodu pocztowego i miasta
        postal_code, city = None, None
        if address_div:
            address_parts = address_div.text.strip().split(',')
            if len(address_parts) > 1:
                # Zakładamy, że druga część adresu zawiera kod pocztowy i miasto
                location_info = address_parts[1].strip()
                postal_code, city = location_info.split(" ", 1) if " " in location_info else (None, location_info)

        # Sprawdzenie, czy numer telefonu istnieje
        phone_element = station.find("a")
        phone = phone_element.text.strip() if phone_element else "Brak telefonu"

        # Dodawanie danych do listy
        all_data.append({
            "Name": name,
            "Address": address,
            "Postal_Code": postal_code,
            "City": city,
            "Phone": phone
        })

# Tworzenie DataFrame z zebranych danych
df = pd.DataFrame(all_data)
print(df)

# Zapis do pliku CSV
df.to_csv("stations_malopolskie.csv", index=False)
```

```python
# Definiujemy puste kolumny dla szerokości i długości geograficznej
df["Latitude"] = None
df["Longitude"] = None

for index, row in df.iterrows():
    # Weź adres z kolumny Address
    address = row["Address"]
    
    # Zbuduj URL API
    url = f"https://api.geoapify.com/v1/geocode/search?text={address}&limit=1&apiKey={apiKey}"
    
    # Wyślij żądanie do API
    response = requests.get(url)
    
    # Sprawdź status odpowiedzi
    if response.status_code == 200:
        # Parsuj dane JSON z odpowiedzi
        data = response.json()
        
        # Sprawdź, czy są dostępne wyniki
        if data["features"]:
            # Wyciągnij szerokość i długość geograficzną z pierwszego wyniku
            result = data["features"][0]
            latitude = result["geometry"]["coordinates"][1]
            longitude = result["geometry"]["coordinates"][0]
            
            # Zapisz wartości do DataFrame
            df.at[index, "Latitude"] = latitude
            df.at[index, "Longitude"] = longitude
            print(f"{address} -> Latitude: {latitude}, Longitude: {longitude}")
        else:
            print(f"Brak wyników dla adresu: {address}")
    else:
        print(f"Żądanie dla adresu '{address}' nie powiodło się z kodem statusu: {response.status_code}")
    
    # Dodaj opóźnienie, aby nie przekroczyć limitu żądań API
    time.sleep(1)  # Możesz dostosować wartość, jeśli to konieczne

# Wyświetl zaktualizowany DataFrame
print(df)
```
